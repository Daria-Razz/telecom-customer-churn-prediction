## Exploratory Data Analysis and Preprocessing

### Contract Information

```python
# Converting the column names of the dataframe to snake case
contract_data.columns = [re.sub(r'(?<!^)(?=[A-Z])', '_', i).lower() for i in contract_data.columns] 
contract_data = contract_data.rename(columns={'customer_i_d':'customer_id'})

# Verifying the updated column names
contract_data.columns
```

```python
# Checking for non-numeric values in total_charges
contract_data.loc[
    contract_data['total_charges'].str.contains(' |,|[a-z]|[A-Z]|[а-я]|[А-Я]'), 
    'total_charges'
].unique()
```

This output shows that there are empty strings in the total_charges column. We can check if these users have monthly_charges as well as the contract duration. Based on this information, we can fill in the total_charges column.

```python
# Filter rows where total_charges is equal to " "
missing_total_charges = contract_data[contract_data['total_charges'] == " "]
print(missing_total_charges)
```

From the provided data, we can draw a few important conclusions:
- All rows where total_charges is equal to " " have a begin_date of 2020-02-01. This may indicate that the data was recorded on the day the contract started, and in fact, services had not yet been rendered.
- Consequently, at the time of the record, total_charges had not yet accumulated, which led to the absence of values in this column. Given that the contracts started and are set to end on the same day, it is logical to assume that services were either not provided or were minimally provided. Therefore, total_charges should either be zero or close to zero.
- Thus, empty values in total_charges can reasonably be replaced with 0. After replacing empty values in total_charges with 0, the entire column should be converted to the float data type.

```python
# Converting total_charges to numeric type in contract_data
contract_data.loc[(contract_data['total_charges'] == " "), 'total_charges'] = 0
contract_data['total_charges'] = pd.to_numeric(contract_data['total_charges'], errors='coerce')
```

```python
# Checking the dataframe for duplicates
contract_data.duplicated().sum()
```

#### Feature Engineering

The features BeginDate and EndDate provide information about the start and end dates of the contract. It is important to note that if "No" is specified in EndDate, it means that the contract was active as of 2020-02-01, the reference date for this dataset.  
  
Since these features are tied to specific time frames, using them directly in the model could lead to prediction errors. Therefore, before removing these features, we will create new ones:
- The feature active_contract will indicate the current status of the contract:  
    - 0 if the contract is active (value "No" in EndDate).  
    - 1 if the contract is terminated (a specific date is provided in EndDate).  
- The feature duration_contract will reflect the duration of the contract in days.  
  
Replacing values in EndDate:
- To correctly calculate the contract duration, the value "No" in EndDate will be replaced with the date 2020-02-01.
  
After creating these new features, BeginDate and EndDate will be removed from the dataset to avoid potential data leakage and errors during model training.

```python
# Creating the feature active_contract
contract_data['active_contract'] = 1
contract_data.loc[contract_data['end_date'] == 'No', 'active_contract'] = 0
```

```python
# Converting the data types of begin_date and end_date
contract_data['begin_date'] = pd.to_datetime(
    contract_data['begin_date'], 
    format='%Y-%m-%d'
)

contract_data.loc[contract_data['end_date'] == 'No', 'end_date'] = '2020-02-01'
contract_data['end_date'] = pd.to_datetime(
    contract_data['end_date'], 
    format='%Y-%m-%d'
)
```

```python
# Creating the feature duration_contract
contract_data['duration_contract'] = (contract_data['end_date'] - contract_data['begin_date']).dt.components.days
```

```python
# Dropping the features begin_date and end_date
contract_data = contract_data.drop(['begin_date', 'end_date'], axis=1)
```

## Merging Datasets

The next step is to merge all the dataframes using the key. In our case, the key is the customer_id column.

```python
print('Dimensions of the contract_data dataframe:', contract_data.shape)
print('Dimensions of the personal_data dataframe:', personal_data.shape)
print('Dimensions of the internet_data dataframe:', internet_data.shape)
print('Dimensions of the phone_data dataframe:', phone_data.shape)
```

The dataframes contract_new and personal_new contain key information about contracts and customers. Since both dataframes have the same number of records, this indicates a complete match of customer identifiers in these datasets. Therefore, they can be merged into a single dataframe merged_data using an inner join.  
The other dataframes contain fewer records, which means they do not cover all customers. Thus, they should be merged with the main dataframe merged_data using a left join to retain information for all customers.

```python
# Merging contract_data with personal_data
merged_data = contract_data.merge(personal_data, on='customer_id', how='inner')

# Merging with internet_data
merged_data = merged_data.merge(internet_data, on='customer_id', how='left')

# Merging with phone_data
merged_data = merged_data.merge(phone_data, on='customer_id', how='left')
```

```python
# Removing the customer identifier 'customer_id' from the dataset before correlation analysis
corr_analysis = merged_data.drop(columns=['customer_id'])
interval_cols = numerical_features

# Performing correlation analysis using phik
phik_overview = corr_analysis.phik_matrix(interval_cols=interval_cols)

# Plotting the heatmap of correlations with numerical values
plt.figure(figsize=(15, 10))
sns.heatmap(phik_overview, annot=True, cmap='coolwarm', linewidths=.5, fmt=".2f")
plt.title('Correlation Analysis Using PhiK')
plt.show()
```

The PhiK correlation heatmap shows the relationships between various features in the dataset. The analysis revealed the following key points:

**Strong Correlations:**

- internet_service and monthly_charges: A correlation of 0.92 is extremely high, indicating that these two features almost completely duplicate information.
- monthly_charges and total_charges: A correlation of 0.71 indicates a significant relationship. Both features reflect aspects of customer expenses but in different contexts (monthly vs. total charges). Including both in the model may lead to some multicollinearity.
- duration_contract and type: A correlation of 0.63 shows that contract type and its duration are closely related. This is logical, as the contract type (e.g., monthly, yearly) directly influences its duration.
- streaming_tv and streaming_movies: A correlation of 0.74 suggests a high likelihood that customers using one of these services also use the other.  
  
**Correlations with the Target Variable active_contract:**
- duration_contract: A correlation of 0.30 indicates that longer contracts are associated with a lower likelihood of churn.
- monthly_charges: A correlation of 0.21 suggests that higher monthly payments may increase the risk of churn.
- total_charges: A correlation of 0.22 also shows a relationship with the target variable, but this is already reflected in monthly_charges.
   
**Weak or Insignificant Correlations:**
- Features such as gender, partner, and dependents show low correlations with both the target variable and other features, indicating their low significance in the context of this task.  
  
**Multicollinearity Recommendations:**
- Exclude internet_service due to its 0.92 correlation with monthly_charges. Since monthly_charges more clearly reflects current customer expenses, it is preferable to keep.
- Exclude total_charges due to its 0.71 correlation with monthly_charges. Since both features are closely related, it makes sense to keep only one. monthly_charges may be preferable as it more clearly reflects current customer expenses. Additionally, total_charges contains possible outliers.
- Exclude type due to its 0.63 correlation with duration_contract. Since duration_contract provides more precise information about the contract length, it is preferable to keep.
- Exclude one of the features: streaming_tv or streaming_movies: A correlation of 0.74 between these features indicates a strong relationship. We will exclude streaming_movies.
- Exclude the features gender, partner, and dependents due to their low correlation with the target variable and other features. These features are unlikely to significantly impact customer churn prediction.

## Machine Learning Model Training 

### Logistic Regression

```python
logistic_preprocessor = make_column_transformer(
    (SimpleImputer(strategy='constant', fill_value=0), numerical_features),
    (OneHotEncoder(drop='first'), categorical_features)
)
```

```python
logistic_pipeline = Pipeline(steps=[
    ('preprocessor', logistic_preprocessor),
    ('classifier', LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced'))
])
```

```python
# Hyperparameter tuning for logistic regression
logistic_params = {
    'classifier__C': [0.1, 1.0, 10.0],
    'classifier__solver': ['lbfgs', 'liblinear']
}
```

```python
logistic_grid = GridSearchCV(logistic_pipeline, logistic_params, cv=5, scoring='roc_auc', n_jobs=-1)
logistic_grid.fit(X_train, y_train)
```

```python
print("Best hyperparameters for Logistic Regression:")
print(logistic_grid.best_params_)
```

### Random Forest

```python
forest_preprocessor = make_column_transformer(
    (SimpleImputer(strategy='median'), numerical_features),
    (OrdinalEncoder(), categorical_features)
)
```

```python
forest_pipeline = Pipeline(steps=[
    ('preprocessor', forest_preprocessor),
    ('classifier', RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced'))
])
```

```python
# Hyperparameter tuning for Random Forest
forest_params = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20, 30],
    'classifier__min_samples_split': [2, 5, 10]
}
```

```python
forest_grid = GridSearchCV(forest_pipeline, forest_params, cv=5, scoring='roc_auc', n_jobs=-1)
forest_grid.fit(X_train, y_train)
```

```python
print("Best hyperparameters for Random Forest:")
print(forest_grid.best_params_)
```

```python
categorical_features = list(categorical_features)
catboost_model = CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)
```

```python
catboost_params = {
    'iterations': [100, 300, 500],
    'depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.8, 1.0]
}
```

```python
catboost_grid = GridSearchCV(catboost_model, catboost_params, cv=5, scoring='roc_auc', n_jobs=-1, verbose=2)
catboost_grid.fit(X_train, y_train, cat_features=categorical_features)
```

```python
print("Best hyperparameters for CatBoost:")
print(catboost_grid.best_params_)
```

## Model Selection

```python
# Evaluating the best models
logistic_best_score = logistic_grid.best_score_
forest_best_score = forest_grid.best_score_
catboost_best_score = catboost_grid.best_score_
```

```python
print(f"Best ROC-AUC on cross-validation for Logistic Regression: {logistic_best_score:.4f}")
print(f"Best ROC-AUC on cross-validation for Random Forest: {forest_best_score:.4f}")
print(f"Best ROC-AUC on cross-validation for CatBoost: {catboost_best_score:.4f}")
```

```python
if max(logistic_best_score, forest_best_score, catboost_best_score) >= 0.85:
    if catboost_best_score >= max(logistic_best_score, forest_best_score):
        best_model = catboost_grid.best_estimator_
        best_model_name = "CatBoost"
        best_model_score = catboost_best_score
    elif forest_best_score >= max(logistic_best_score, catboost_best_score):
        best_model = forest_grid.best_estimator_
        best_model_name = "Random Forest"
        best_model_score = forest_best_score
    else:
        best_model = logistic_grid.best_estimator_
        best_model_name = "Logistic Regression"
        best_model_score = logistic_best_score
else:
    print("None of the models reached the ROC-AUC threshold of >= 0.85 on cross-validation.")
    best_model = None
    best_model_name = None
    best_model_score = None
```

```python
if best_model is not None:
    y_pred_test = best_model.predict(X_test)
    y_pred_proba_test = best_model.predict_proba(X_test)[:, 1]

    test_roc_auc = roc_auc_score(y_test, y_pred_proba_test)
    print(f"ROC-AUC on the test set for {best_model_name}: {test_roc_auc:.4f}")
    
    if test_roc_auc >= 0.97:
        print(f"Warning: the {best_model_name} model may be overfitted. Check for data leakage.")
    
    print(f"Classification report for {best_model_name}:")
    print(classification_report(y_test, y_pred_test))
    
    # ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_test)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (area = {test_roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
```

The model performs well in predicting customers who stay but has a slight reduction in recall for those who churn. This may indicate that the model is cautious and prefers to minimize false positives.
